## Augmentation avec le CRAC de 2020

```python
import pandas as pd 
df = pd.read_csv("./graph_3_years/hal_depot_2018_2020.csv")
df
```

```python
# comparer l'évol de 2020 par rapport à 2019
df["file_evol_from_last_year"] = df["file"].pct_change(periods = 12) * 100
df[24:36]
```

```python
# calculer l'augmentatino moyenne des dépôt pour 2020
## attention : retirer tous les mois où des versements auto sont présents (prodinra)
# mois à inclure janvier, mars, avril, juillet, aout, sept, oct, nov, dec, 
include_month = ["2020-1", "2020-3", "2020-4", "2020-7", "2020-8", "2020-9", "2020-10", "2020-11", "2020-12"]
df_2020 = df[ df["year-month"].isin(include_month)]
aug_moy_2020 = df_2020["file_evol_from_last_year"].mean()
aug_moy_2020
```

```python
print("augmentation moyenne de 2020 par rapport a 2019 pour les dépôts de fichier", round(aug_moy_2020), "%")
```

```python
# calc de la moyenne de l'augmentation sur nov & dec
aug_2020_nov_dec = df.loc[ df["year-month"].isin(["2020-11", "2020-12"]), "file_evol_from_last_year"].mean()
aug_2020_nov_dec
```

```python
print("augmentation moyenne de nov. 2020 comparé à nov.2019 pour les fihciers ", round(aug_2020_nov_dec), "%")
```
